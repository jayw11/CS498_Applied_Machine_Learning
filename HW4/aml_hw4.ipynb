```
import keras
from keras.datasets import mnist
from sklearn.cluster import KMeans
import numpy as np
import random

batch_size = 128  
num_classes = 10
epochs = 12
img_rows, img_cols = 28, 28
index = [0,6,12,18]

(x_train, y_train), (x_test, y_test) = mnist.load_data()
#print(np.shape(x_train))
```

```
sample_6000 = []
#choose 6k random index out of 60k
randomlist = random.sample(range(0, 60000), 6000)
for idx in randomlist:
  #for each of the 6k images, choose a random patch by choosing a 
  #random x and y which is the top-left index of a 10*10 patch
  x_idx = random.choice(index)
  y_idx = random.choice(index)
  sample = [[0 for i in range(10)] for j in range(10)]
  for i in range(x_idx, x_idx+10):
    for j in range(y_idx, y_idx+10):
      sample[i-x_idx][j-y_idx] = x_train[idx][i][j]
  sample_6000.append(sample)
print(sample_6000)

d2_sample_6000 = np.array(sample_6000).reshape((6000,100))
print(d2_sample_6000)
kmeans = KMeans(n_clusters=50).fit(d2_sample_6000)
#print(len(kmeans.labels_))
```

```
sample_60000 = []
for idx in range(60000):
  x_idx = random.choice(index)
  y_idx = random.choice(index)
  sample = [[0]*10]*10
  for i in range(x_idx, x_idx+10):
    for j in range(y_idx, y_idx+10):
      sample[i-x_idx][j-y_idx] = x_train[idx][i][j]
  sample_60000.append(sample)

d2_sample_60000 = np.array(sample_60000).reshape((60000,100))
labels_60000_50 = kmeans.predict(d2_sample_60000)
print(len(set(labels_60000_50)))

datasets = [0]*50
for i in range(60000):
  if datasets[labels_60000_50[i]] == 0:
    datasets[labels_60000_50[i]] = [d2_sample_60000[i]]
  else:
    datasets[labels_60000_50[i]].append(d2_sample_60000[i])
print(len(datasets))
```

```
kmeans_50 = [0]*50
for i in range(50):
  kmeans_50[i] = KMeans(n_clusters=50).fit(datasets[i])
print(len(kmeans_50),len(set(kmeans_50[0].labels_)))
```

```
labels_60000_2500 = []
for i, kmeans_model in enumerate(kmeans_50):
  for label in kmeans_model.labels_:
    labels_60000_2500.append(i*50+label)
print(len(labels_60000_2500))

patch_60000 = []
for i in range(50):
  patch_60000.extend(datasets[i])
print(len(patch_60000))
```
