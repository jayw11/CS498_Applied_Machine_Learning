#em steps:

```
def em(cluster_num, image, cov):
    converge_level = 0.1
    tmp_array = np.reshape(image, (len(image) * len(image[0]), 3))
    kmeans = KMeans(n_clusters=cluster_num, max_iter=1).fit(tmp_array)
    cur_means = kmeans.cluster_centers_
    pi = np.array([len(np.where(kmeans.labels_ == x)[0]) for x in np.unique(kmeans.labels_)]).astype(
        'float32') / len(kmeans.labels_)
    maps = np.zeros((len(image), len(image[0]), cluster_num))
    prev_means = np.copy(cur_means)

    for i in range(20):
        for cluser_id in range(len(cur_means)):
            mean_f = means.astype(float)
            up, down= np.zeros((3,)), 0
            for i in range(len(image)):
                for j in range(len(image[0])):
                    pixel_f = image[i][j].astype(float)
                    val = [-0.5 * np.dot(np.dot((pixel_f - mean_f[i]).T, cov), (pixel_f - mean_f[i])) for i in
                           range(cluster_num)]
                    up2, down2 = np.exp(val[cluser_id]) * pi[cluser_id], np.dot(np.exp(val), pi)
                    wij = up2 / down2
                    maps[i][j][cluser_id] = wij
                    up += image[i][j] * wij
                    down += wij
            cur_means[cluser_id] = (up / down)
            pi[cluser_id] = (down / (len(image) * len(image[0])))
        dis = sum([euclidean(cur_means[i], prev_means[i]) for l in range(len(cur_means))])
        if dis < converge_level:  return cur_means, maps
        # cluster_n = {}
        # for i in range(len(image)):
        #     for j in range(len(image[0])):
        #         val_max = np.argmax(maps[i][j])
        #         cluster_n[val_max] = cluster_n.get(val_max, 0) + 1
        # print("clusters distribution: ", cluster_n, " diff: ", diff)
    return cur_means, maps
```
